```python
% matplotlib inline
% reload_ext autoreload
% autoreload 2

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.cm import ScalarMappable
from matplotlib.colors import Normalize

from system import RoomEnv, cost, reward
from system import ALPHA, GAMMA, N, CCENTERS, INTERCEPT, SAMPLES
import qlearn

NCLUSTERS = len(CCENTERS)
# Custom color map and legend for plots
cmap = plt.get_cmap('Set1', lut=NCLUSTERS)
rgb_conv = ScalarMappable(cmap=cmap, norm=Normalize(0, NCLUSTERS))
legend_patches = [Patch(color=rgb_conv.to_rgba(i),
                        label='Cluster{}'.format(i)) for i in range(NCLUSTERS)]
```

# Energy Consumption Model

A plot of the learned energy consumption model. It is used as a basis of the cost and reward for taking actions in the environment. The energy consumption behavior is modelled using Support Vector Regression after clustering data points.

```python
env = RoomEnv()
low, high = env.action_space.low[0], env.action_space.high[0]
dt_range = np.linspace(low, high, 50)
fig, ax = plt.subplots(figsize=(8, 6))

for i, cluster in enumerate(CCENTERS):
    energy = []
    for dt in dt_range:
        state = cluster[:]
        state[-1] = dt
        energy.append(cost(ALPHA[i], SAMPLES[i], state,\
                           GAMMA[i], N[i], INTERCEPT[i]))
    ax.scatter(dt_range, energy, c=[i]*len(energy), cmap=cmap,
                vmin=0, vmax=NCLUSTERS-1)

fig.legend(handles=legend_patches)
plt.show()
```

# Derived Reward Function

The reward is the energy consumption/cost of a state as predicted by the closest cluster.

```python
env = RoomEnv()
low, high = env.action_space.low[0], env.action_space.high[0]
dt_range = np.linspace(low, high, 50)
fig, ax = plt.subplots(figsize=(8, 6))

for i, cluster in enumerate(CCENTERS):
    rewards = []
    cindices = []
    for dt in dt_range:
        state = cluster[:]
        state[-1] = dt
        r, c = reward(None, None, state)
        rewards.append(r)
        cindices.append(c)
    ax.scatter(dt_range, rewards, c=cindices, cmap=cmap,
                vmin=0, vmax=NCLUSTERS-1)

fig.text(x=0, y=0, s=('Rewards of different Discharge Temperature '
                        'settings with all other state variables '
                        'equal to cluster centers.'), wrap=True)
fig.legend(handles=legend_patches)
plt.show()
```

```python
env = RoomEnv()
low, high = env.action_space.low[0], env.action_space.high[0]
dt_range = np.linspace(low, high, 50)
fig, ax = plt.subplots(figsize=(8, 6))

for i, state in enumerate([env.observation_space.sample() for _ in range(10)]):
    rewards = []
    cindices = []
    for dt in dt_range:
        state[-1] = dt
        r, c = reward(None, None, state)
        rewards.append(r)
        cindices.append(c)
    ax.scatter(dt_range, rewards, c=cindices, cmap=cmap,
                vmin=0, vmax=NCLUSTERS-1)

fig.text(x=0, y=0, s=('Rewards of different Discharge Temperature '
                              'settings for states sampled randomly.'))
fig.legend(handles=legend_patches)
plt.show()
```

# Control

```python
from qlearn.agent import QAgent, NStepSarsaAgent, NStepTDAgent
from qlearn.approximator import Tabular, Neural
from qlearn.helpers.parameters import LinearSchedule

env = RoomEnv()

# Tabular value function parameters
t_dims = (2, 2, 2, 2, 50)
# high and low limits for tabular value function
t_l = np.asarray([*env.observation_space.low, env.observation_space.low[-1]])
t_h = np.asarray([*env.observation_space.high, env.observation_space.high[-1]])

# Learning parameters
episodes = 300
lrate = LinearSchedule(0.01, 0.001, episodes)
epsilon=LinearSchedule(0., 0.9, episodes)
discount = LinearSchedule(0.9, 0.9, episodes)
```

## Dumb agent

```python
env = RoomEnv(random_state=0)
value = Tabular(dims=t_dims, low=t_l, high=t_h, lrate=0., random_state=0)
agent = QAgent(env=env, value_function=value, random_state=0)
r = agent.learn(episodes=episodes, epsilon=0, discount=0)

fig, ax = plt.subplots()
ax.scatter(np.linspace(1, episodes, episodes), r)
plt.show()
print(np.mean(r[-100:]))
```

## Q-Learning

```python
env = RoomEnv(random_state=0)
value = Tabular(dims=t_dims, low=t_l, high=t_h, lrate=lrate, random_state=0)
# value = Neural((4,4), learning_rate_init=0.01, learning_rate='adaptive', random_state=0)
agent = QAgent(env=env, value_function=value, random_state=0)
r = agent.learn(episodes=episodes, epsilon=epsilon, discount=discount,
                memsize=1, batchsize=1, steps=1)

fig, ax = plt.subplots()
ax.scatter(np.linspace(1, episodes, episodes), r)
plt.show()
print(np.mean(r[-100:]))
```
